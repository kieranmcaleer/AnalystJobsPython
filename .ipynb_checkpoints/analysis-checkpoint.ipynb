```python
#bring the data in
import pandas as pd
import numpy as np
import seaborn as sns
import re
import matplotlib.pyplot as plt
job_df = pd.read_csv('archive/DataAnalyst.csv')
```

```python
#filter out the data that is not needed
job_df = job_df[job_df['Job Title'].str.contains('Data')]
job_df = job_df[job_df['Job Title'].str.contains('Analyst')]
```

```python
#all rows contain data analyst now we can start breaking down the jobs according to what how many times they mention or if they mention a specific language
added_df = job_df[job_df['Job Description'].str.contains("tableau" ,na=False, flags=re.IGNORECASE)]
added_df
count_row = added_df.shape[0]
count_row
```

```python

#SQL
added_df["sql"] = added_df['Job Description'].str.contains("sql" ,na=False, flags=re.IGNORECASE)
#Python
added_df["python"] = added_df['Job Description'].str.contains("python" ,na=False, flags=re.IGNORECASE)
#R
added_df["sql"] = added_df['Job Description'].str.contains("sql" ,na=False, flags=re.IGNORECASE)
#Excel
added_df["excel"] = added_df['Job Description'].str.contains("excel" ,na=False, flags=re.IGNORECASE)
#Tableau
added_df["tableau"] = added_df['Job Description'].str.contains("tableau" ,na=False, flags=re.IGNORECASE)
#PowerBI
added_df["powerbi"] = added_df['Job Description'].str.contains("powerbi" ,na=False, flags=re.IGNORECASE)
#Machine Learning or ML
added_df["ml"] = added_df['Job Description'].str.contains("ml" or "machine learning" ,na=False, flags=re.IGNORECASE)
#Jupyter
added_df["jupyter"] = added_df['Job Description'].str.contains("jupyter" ,na=False, flags=re.IGNORECASE)
#Spark
added_df["spark"] = added_df['Job Description'].str.contains("spark" ,na=False, flags=re.IGNORECASE)

```
